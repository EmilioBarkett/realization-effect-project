```python
df = pd.read_csv("../results/results.csv")
```

```python
df.head()
df.shape
```

```python
df["wager"] = pd.to_numeric(df["wager"], errors="coerce")
df = df.dropna(subset=["wager"])
```

```python
trial_id	condition	outcome_type	amount	run_number	model	temperature	prompt_version	prompt_text	response_text	parsed_wager	log_wager	valid	refusal_flag	parse_error_type	timestamp	request_id
```

```python
df["wager"] = pd.to_numeric(df["parsed_wager"], errors="coerce")
df = df.dropna(subset=["wager"])
```

```python
df.groupby("condition")["wager"].mean().sort_values()
```

```python
df["is_realized"] = df["outcome_type"].eq("realized").astype(int)
df["is_loss"] = (df["amount"] < 0).astype(int)
df["is_gain"] = (df["amount"] > 0).astype(int)
df["abs_amount"] = df["amount"].abs()
```

```python
import statsmodels.formula.api as smf

model = smf.ols(
    "wager ~ is_loss * is_realized + abs_amount",
    data=df
).fit()

print(model.summary())
```
